```python
sns.countplot(data = df , x = "Sex")
plt.title("Distrubution by Gender")
plt.show()
```

```python
sns.histplot(data = df , x = "Age", bins = 10 , kde = True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data = df , x = "Height" , bins = 10 , kde = True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data = df , x = "Weight" , bins = 10 , kde = True)
plt.title("Weight Distribution")
plt.show()
```

```python
sns.countplot(data = df, x = "Medal", hue = "Sex")
plt.title("Medals Distribution by Sex")
plt.show()
```

```python
sns.countplot(data = df, x = "Medal", hue = "Year")
plt.title("Medals Distribution by Sex")
plt.show()
```

```python
Year_Avg_Age = df.groupby("Year")["Age"].mean()
print(Year_Avg_Age)
```

```python
sport_median_height = df.groupby("Sport")["Height"].median()
print(sport_median_height)
```

```python
sport_median_height[sport_median_height == 164.0]
```

```python
country_gender_count = df.groupby(["NOC","Sex"])["ID"].count()
print(country_gender_count)
```

```python
country_medal_count = df[df["Medal"] == "Gold"].groupby("NOC")["Medal"].count()
print(country_medal_count.max())
```

```python
country_medal_count[country_medal_count == 747]
```

```python
top_sports = df.groupby('Sport')['Medal'].count().sort_values(ascending=False).head(10)

top_sports.plot(kind='barh', figsize=(10, 6))
plt.title('Top 10 Sports by Medal Count')
plt.xlabel('Number of Medals')
plt.show()
```

```python
Sports_gender_Avg_weight = df.groupby(["Sport", "Sex"])["Weight"].mean()
print(Sports_gender_Avg_weight["Wrestling"]["F"]) 
```

```python
Year_participant_count = df.groupby("Year")["ID"].nunique()
Year_participant_count.plot(kind = "bar")
plt.title("Number of Participant Over the Year")
plt.xlabel("Year")
plt.ylabel("Number of participants")
plt.show()
```

```python
Country_Avg_Age = df.groupby('NOC')['Age'].mean().sort_values(ascending=False).tail(10)
Country_Avg_Age.plot(kind="bar", figsize=(10, 6))
plt.title("Top 10 Countries with Lowest Average Age of Participants")
plt.xlabel("Country")
plt.ylabel("Average Age")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data = df , x = "Season" , y = "Age")
plt.title("Distribution of Age by Season")
plt.xlabel("Season")
plt.ylabel("Age")
plt.show()
```

```python
sns.violinplot(data = df , x = "Medal" , y = "Height", palette = "Set2")
plt.title("Distribution of Heights by Medal")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.show()
```
